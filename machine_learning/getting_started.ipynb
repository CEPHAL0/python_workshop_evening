# Python

Validate the installation of python in your laptops using the following command in terminal

1. MacOS command
```shell
python3 --version
```

2. Windows Command
```shell
python --version
```

# Jupyter Lab

Validate the installation of jupyter notebook in your laptop using the following command in your terminal

```shell
jupyter lab
```

# Python Interpreter Check

Check the python interpreter health in your laptop using the following command in your terminal


1. MacOS
```shell
python3

print("Hello World")

exit()
```

2. Windows
```shell
python

print("Hello World")

exit()
```

# Install PIP (Python Interpreter Package)

Install PIP for handling packages and bundling packages in venv

1. MacOS
```shell
python3 -m ensurepip --upgrade
```

2. Windows
```shell
python -m ensurepip --upgrade
```
