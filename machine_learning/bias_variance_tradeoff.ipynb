<h1 style="text-align:center;">Real Model Example</h1>

| Model | Bias | Variance |
|-----|-----|-----|
| Linear Regression | High | Low |
| Decision Tree | Low | High |
| Random Forest | Medium | Medium |
| Neural Network (small) | Medium | Medium |
| Neural Network (big) | Low | High |

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 1, 20)[:, None]
X
```

```python
y = np.sin(2*np.pi*X).ravel() + np.random.randn(20) * 0.2
```

```python
y
```

```python
x_plot = np.linspace(0, 1, 100)[:, None]
```

```python
# Try different model complexities
degrees = [1, 2, 3, 5, 7, 9, 12]
for degree in degrees:
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)
    y_plot = model.predict(x_plot)
    plt.plot(x_plot, y_plot, label=f"Degree {degree}")

plt.legend()
plt.title("Bias Variance Tradeoff")
plt.show()
```
