```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# Generate an artificial data
X = np.linspace(0, 10, 100).reshape(-1, 1)
X
```

```python
# y = 1.5x^2 + 2x + 3
y = 3 + 2*X + 1.5*X**2 + np.random.randn(100, 1)*5
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
X_poly = poly.fit_transform(X)
X_poly
```

```python
X_poly.shape
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_poly)
```

```python
plt.scatter(X, y, label='Data')
plt.plot(X, y_pred, linewidth=3, label='Polynomial Fit')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("data/Salary_dataset.csv")
```

```python
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)
```

```python
X_train_poly
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test_poly)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_pred, y_test)
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
rmse = mean_squared_error(y_pred, y_test) ** 1/2
rmse
```

```python
r2_score(y_pred, y_test)
```

```python
degree = 3

initial_data = 6 # Predict the salary of n year experience
data = []
for i in range(0, degree + 1):
    data.append(initial_data ** i)

model.predict([data])
```
