```python
import numpy as np
```

```python
numbers_1 = [10, 20, 30, 40, 50]
numbers_2 = [40, 50, 60, 70]

display(type(numbers_1))
type(numbers_2)
```

```python
# Converting simple python list to numpy array
numbers_1 = np.array(numbers_1)
numbers_2 = np.array(numbers_2)
```

```python
display(type(numbers_1))
type(numbers_2)
```

```python
# 1 dimensional array
display(numbers_1.shape)

# Shape of array
numbers_2.shape
```

```python
# Dimension of array

display(numbers_1.ndim)
numbers_2.ndim
```

```python
numbers_1.size
```
