```python
# Categorization using loc
df.loc[df['age'] < 11, 'age_group'] = 'Kid'
df.loc[(df['age'] >= 11) & (df['age'] < 20), 'age_group'] = 'Teen'
df.loc[(df['age'] >= 20) & (df['age'] < 31), 'age_group'] = 'Early Adult'
df.loc[(df['age'] >= 31) & (df['age'] < 51), 'age_group'] = 'Adult'
df.loc[(df['age'] >= 51) & (df['age'] < 81), 'age_group'] = 'Elderly'
df.loc[df['age'] >= 81, 'age_group'] = 'Old'
```

```python
# Categorization using np
conditions = [
    df['age'] < 11,
    (df['age'] >= 11) & (df['age'] < 20),
    (df['age'] >= 20) & (df['age'] < 31),
    (df['age'] >= 31) & (df['age'] < 51),
    (df['age'] >= 51) & (df['age'] < 81),
    df['age'] >= 81
]

mappings = ['Kid', 'Teen', 'Early Adult', 'Adult', 'Elderly', 'Old']

df['age_group_np'] = np.select(conditions, mappings, default='Unknown')
```

```python
# Defining custom function to apply
def get_systolic(value):
    splitted_values = value.split("/")
    systolic = splitted_values[0]
    return systolic

df['systolic'] = df['blood_pressure'].apply(get_systolic)


# using lambda function
df['systolic'] = df['blood_pressure'].apply(lambda value: value.split('/')[0])
```

```python
df['diastolic'] = df['blood_pressure'].apply(lambda value: value.split('/')[1])
```

```python
df
```

```python
df[['systolic', 'diastolic']] = df[['systolic', 'diastolic']].astype(int)
```

```python
df.dtypes
```

```python
df.loc[df['systolic'] > 120, 'systolic_ind'] = 'High'
df.loc[df['systolic'] < 110, 'systolic_ind'] = 'Low'
df.loc[(df['systolic'] <= 120) & (df['systolic'] >= 110), 'systolic_ind'] = 'Normal'
```

```python
df.loc[df['diastolic'] > 80, 'diastolic_ind'] = 'High'
df.loc[df['diastolic'] < 70, 'diastolic_ind'] = 'Low'
df.loc[(df['diastolic'] <= 80) & (df['diastolic'] >= 70), 'diastolic_ind'] = 'Normal'
```

```python
df.head()
```

```python
df.loc[df['systolic_ind'] != 'Normal', 'final_ind'] = df['systolic_ind']
df.loc[df['systolic_ind'] == 'Normal', 'final_ind'] = df['diastolic_ind']
```
