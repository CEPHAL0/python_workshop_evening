```python
# Task: Generate a random array of integers of 4 dimension and Generate another random array 
# of floats of 4 dimension and then perform operations: add, multiply, division
# Reshape the sum array to 2 dimensions, product array to 3 dimensions and division array to 1 dimension
# Sort sum array in descending order, product array in ascending order, division array in descending order
# Reshape sorted division array to 2 dimensions and concat the sorted sum array and reshaped division array
```

```python
integer_arr = np.random.randint(10, 100, size=(3, 2, 2, 5))
float_arr = 100 * np.random.rand(3, 2, 2, 5)

display(integer_arr)
float_arr
```

```python
addition = integer_arr + float_arr
multiply = integer_arr * float_arr
division = integer_arr / float_arr

addition, multiply, division
```

```python
addition = np.reshape(addition, (6, 10))
multiply = np.reshape(multiply, (4, 5, 3))
division = np.reshape(division, (60,))

addition, multiply, division
```

```python
addition.shape, multiply.shape, division.shape
```

```python
addition = np.sort(addition, axis=1)[:, ::-1]
multiply = np.sort(multiply, axis=2)
division = np.sort(division)[::-1]

addition, multiply, division
```

```python
division = np.reshape(division, (6, 10))
division
```

```python
concat_arr = np.concat((addition, division), axis=1)
```
