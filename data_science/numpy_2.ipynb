```python
# arange - Automatic Range
import numpy as np
```

```python
# Start, Stop, Step
array_1 = np.arange(10)
```

```python
array_1
```

```python
array_2 = np.arange(10, 41, 5)
```

```python
array_2
```

```python
array_3 = np.arange(100, 10, -10)
array_3
```

```python
# Ordering or Sorting
array_3 = np.sort(array_3)
array_3
```

```python
# List Indexing and Slicing for reverse sorting (descending)

array_3 = array_3[::-1]
array_3
```

```python
# Sort a two dimensional array
array_2d = np.array([[45, 34, 42], [89, 91, 44], [65, 42, 33]])
```

```python
array_2d = np.sort(array_2d, axis=1) # axis=0 for column and axis=1 for row
array_2d
```

```python
array_2d.ndim
```

```python
array_2d.shape
```

```python
array_2d_reverse = array_2d[:, ::-1]
array_2d_reverse
```

```python
# Concatenation - Joining two different arrays

# Parameter is a tuple
arrays_to_concat = (array_2d, array_2d_reverse)

# Pass the tuple as a parameter to the concat method
array_concat = np.concat(arrays_to_concat)
```

```python
array_concat
```

```python
# Reshape - To change the dimension
# 1d array [10, 20, 30, 40, 50, 60]
# 2d array [
#     [10, 20, 30],
#     [40, 50, 60]
#     ]

array_1 = np.arange(10, 101, 10)
```

```python
array_1.shape
```

```python
# Convert (10,) to (5,2)
array_converted = np.reshape(array_1, (5, 2))
```

```python
display(array_converted)
array_converted.shape
```
