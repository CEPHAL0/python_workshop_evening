```python
balance = 10000
```

```python
# Function 1 -> Deposit Amount
# Function 2 -> Withdraw Amount
# Function 3 -> Check Balance
# Function Definition
def deposit_amount():
    global balance
    amount = int(input("Enter the amount you want to deposit: "))
    balance += amount
    print(f"You have successfully deposited {amount}, new balance is {balance}")
```

```python
def check_withdraw_eligibility(amount):
    if amount > balance:
        print("Insufficient Balance")
    else:
        print("Sufficient Balance")
```

```python
# Function Definition 2
def withdraw_amount(amt):
    # Function Call 1
    check_withdraw_eligibility(amt)
    balance -= amt
    print(f"You have successfully withdrawed {amt}")
```

```python
# Function Call
deposit_amount()
```
