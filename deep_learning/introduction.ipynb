# Neurons

Also known as perceptrons.

Each neuron has a **bias** and **weight**




Neuron takes an input signal and gives out another output signal.

In between the input signal and output signal, we have a processing part.

The processing function in between is also called as an **activation function**.



# Activation Functions



Collection of multiple neurons in a connected manner is called neural network.

```python
from sklearn.datasets import load_iris
from keras.utils import to_categorical
```

```python
iris = load_iris()
X, y = iris['data'], iris['target']
```

```python
y = to_categorical(y)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
model = Sequential([
    # Input layer
    Dense(10, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    # Hidden Layer 1
    Dense(10, activation='relu'),
    # Output Layer
    Dense(3, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
history = model.fit(
    X_train_scaled, 
    y_train,
    epochs=100,
    batch_size=16,
    validation_data = (X_test_scaled, y_test)
)
```

```python
model.layers[2].get_weights()
```

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
```

```python
loss, accuracy
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Iris Data: Loss vs Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss (Categorical CrossEntropy)")
plt.legend()
plt.show()
```
